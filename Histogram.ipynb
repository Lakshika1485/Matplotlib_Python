# What is a Histogram?
A histogram is a type of bar graph that represents 
the distribution of numerical data by grouping data into ranges (bins). Each bar shows 
the frequency (or count) of data points that fall within a particular bin.

# Function Signature

```python
hist(x, bins=None, range = None, density = False, weights = None, cumulative = False, 
     bottom = None, histtype='bar', align='mid', orientation='Vertical', rwidth= None, log = False, 
    color=None, label = None, stacked = False, data = None)
```

# Why is a Histogram Useful?

1. Shows how data is distributed over intervals.

2. Helps identify the shape of the data distribution (e.g., normal, skewed).

3. Makes it easier to spot outliers, gaps, and peaks.

4. Useful for summarizing large datasets and understanding patterns.

# Attributes of a Histogram
x: Input data (a sequence of numerical values).

bins: Number of bins (intervals) or specific bin edges.

range: Lower and upper range of the bins.

density: If True, the area under the histogram integrates to 1 (used for probability distributions).

weights: Array of weights, same length as x.

cumulative: If True, shows cumulative values.

bottom: Array of values to place the base of each bar.

histtype: Type of histogram ('bar', 'barstacked', 'step', 'stepfilled').

align: Alignment of bars ('left', 'mid', 'right').

orientation: 'vertical' (default) or 'horizontal'.

rwidth: Relative width of the bars.

log: Plot the y-axis on a log scale.

color: Bar color.

label: Used for adding a label (in legends).

stacked: If True, multiple datasets are stacked.

data: Dataset passed as a dictionary or DataFrame.

# When is a Histogram Used?

1. When analyzing frequency distribution of continuous variables.

2. When summarizing large sets of numerical data.

3. When visualizing spread, central tendency, and variability.

4. When identifying data patterns, like skewness or modality.

# Purpose of a Histogram

1. To display the underlying frequency distribution of data.

2. To provide insight into the data's spread, center, and shape.

3. To help detect outliers or unusual data points.

4. To group continuous data into ranges for analysis.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ages = [18,19,21,25,26,26,30,32,38,45,55]
# plt.hist(ages)
plt.hist(ages, bins = 5, edgecolor = 'black')

plt.title("Ages of Respondents")
plt.xlabel('Ages')
plt.ylabel('Total Respondents')
```

```python
bins = [10,20,30,40,50,60]

plt.hist(ages, bins =bins , edgecolor = 'black')

plt.title("Ages of Respondents")
plt.xlabel('Ages')
plt.ylabel('Total Respondents')
```
