# Bar Chart

A bar chart is a graph that presents categorical data with rectangular bars, where the length or height of each bar 
is proportional to the value it represents. 
The bars can be vertical or horizontal, and are used to compare values across categories.

# Function Signature

# Why is a Bar Chart Useful?
Clearly displays comparisons between different categories.

Makes it easy to see which categories have the highest or lowest values.

Handles both small and large data categories effectively.

Supports grouped or stacked bar charts for deeper analysis.



# Attributes of a Bar Chart
x: Sequence of categories (e.g., names, indices).

height: Heights of the bars (i.e., values).

width: Width of the bars (default is 0.8).

bottom: Y-coordinate of the bottom of the bars (used for stacked bars).

align: Aligns bars to the 'center' or 'edge' of x-ticks.

data: Accepts structured input like a dictionary or DataFrame.

color: Fill color of the bars.

edgecolor: Border color of the bars.

linewidth: Width of the edges.

label: Label for each bar (used in legend).

alpha: Transparency of the bars.

tick_label: Names displayed on the axis ticks (if different from x).

title(), xlabel(), ylabel(): Used to add chart title and axis labels (not part of bar() but commonly used with it).



# When is a Bar Chart Used?

1. When you want to compare quantities across discrete categories.

2. When the data involves non-continuous groups (e.g., departments, regions, products).

3. When showing exact values is important.

4. When you need to visualize frequency, count, or total value per category.



# Purpose of a Bar Chart
1. To compare values across different groups.

2. To highlight significant differences in data.

3. To offer a simple and clear view of categorical data distribution.

4. To support decision-making by showing which areas perform better or worse

```python
with open('data.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)

    language_counter = Counter()

    for row in csv_reader:
        language_counter.update(row['LanguagesWorkedWith'].split(';'))

languages = []
popularity = []

for item in language_counter.most_common(10):
    languages.append(item[0])
    popularity.append(item[1])

# plt.bar(languages, popularity) #using this, our bar becomes unreasonably faulty so we will use barh horizontol one for the fault tolerance
plt.barh(languages, popularity)


plt.title("Most popular languages")
plt.xlabel("Number of people work use")
plt.ylabel("Programming Language")

plt.tight_layout()
plt.grid(True)
plt.show()
```

As you can see the bar char is having the most worked language at the bottom so what if we want to see it at top. what are we going to do abt it is :
we can reverse it

```python
with open('data.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)

    language_counter = Counter()

    for row in csv_reader:
        language_counter.update(row['LanguagesWorkedWith'].split(';'))

languages = []
popularity = []

for item in language_counter.most_common(10):
    languages.append(item[0])
    popularity.append(item[1])
# before barh method we can just use reverse method
languages.reverse()
popularity.reverse()

plt.barh(languages, popularity) #barh = bar chart horizontal mein

plt.title("Most popular languages")
plt.xlabel("Number of people work use")

plt.tight_layout()
plt.grid(True)
plt.show()
```
