# What is a Scatter Plot in matplotlib?
A scatter plot is a type of data visualization used to display the relationship between two numerical variables. In matplotlib, the function plt.scatter() is used to create scatter plots.

Each point on the plot represents a single data point with its x and y coordinates.

A scatter plot is a type of plot that uses dots to represent the values of two different variables. Each dot represents a single data point, with its position determined by its value on the x-axis and y-axis. Itâ€™s used to show relationships or correlations between two continuous variables.

# Why is a Scatter Plot Used?
1. Scatter plots are commonly used to:

2. Show correlation or relationships between variables.

3. Detect trends, clusters, and outliers.

4. Visualize distribution of data points.

5. Useful in regression analysis and machine learning to understand feature relationships.

# Important Attributes/Parameters of plt.scatter():

Parameter	Description

x    : 	            X-axis data (array-like)

y    :	            Y-axis data (array-like)

s	 :            Size of the markers (scalar or array)

c	 :            Color of markers (single color or array of colors)

marker	  :      Style of marker (e.g., 'o', 's', '^', '*', etc.)

alpha	  :      Opacity (0: transparent, 1: solid)

cmap	  :      Colormap for coloring by values in c (useful for coloring by category)

edgecolors	 :   Color of marker edges

linewidths	 :   Width of the marker edge

label	     :   Label for the legend

norm	     :   Normalize data values to 0-1 for colormap

vmin, vmax	  :  Bounds for colormap normalization

zorder	      :  Order for layering plots (higher zorder on top)



# When is a Scatter Plot Used?
1. When examining the relationship between two continuous variables.

2. When trying to detect trends, clusters, or patterns.

3. When looking to identify outliers.

4. When performing data correlation and distribution analysis.



# Purpose of a Scatter Plot

1. To visualize the relationship between variables.

2. To identify correlation (positive, negative, or no correlation).

3. To detect data patterns like clusters or gaps.

4. To assist in regression analysis and predictive modeling.



```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = [1, 2, 3, 4, 5]
y = [5, 4, 3, 2, 1]
sizes = [50, 100, 200, 300, 400]
colors = ['red', 'blue', 'green', 'purple', 'orange']

plt.scatter(x, y, s=sizes, c=colors, marker='o', alpha=0.7, edgecolors='black', linewidths=1)
plt.title("Sample Scatter Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

```python
data = pd.read_csv("data4.csv")
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.scatter(view_count, likes,edgecolor = 'black', linewidth = 1, alpha = 0.75)

plt.title('Trending Youtube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()
plt.show()
```

```python
data = pd.read_csv("data4.csv")
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.scatter(view_count, likes,c = ratio, cmap = 'summer', edgecolor = 'black', linewidth = 1, alpha = 0.75)

cbar = plt.colorbar()
cbar.set_label('Like/Dislike Ratio')

plt.xscale('log')
plt.yscale('log')

plt.title('Trending Youtube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()
plt.show()
```
