```python
import matplotlib.pyplot as plt
```

```python
slices = [24.8, 32.2, 19.5, 8.3, 10.2, 3.5, 1.5]
label= ["T.V.", "Mobile", "Video Games", "Crockries", "Tablet", "Laptop", "Printer"]
color = ["green","red", "orange", "yellow" , "grey", "blue" , "brown"]
```

```python
plt.pie(slices, labels = label,colors = color, wedgeprops = {'edgecolor': 'black'})
plt.title("List of sold item in a month")
plt.show()
```

```python
plt.pie(slices, labels = labels, wedgeprops= {'edgecolor': 'black'})
plt.title("Programming langs")

plt.show()
```

explode method: Jab hum pie chart banate hain using matplotlib.pyplot.pie() (yaani plt.pie()), toh explode ek list hoti hai jo batati hai ki kaunsa slice chart se bahar nikaalna hai thoda sa  jisse wo zyada highlight ho jaye

shadow true krne se pie chart ki shadow bn jayegi

startangle = 90 krne se invert 90 hojega, u can change the angle as the circle

autopct ek parameter hai plt.pie() function ka  iska kaam hai ki har slice ke upar percentage ka label dikhaye. Automatically show the percentage on each slice.

auto = automatically
pct = percentage


%1.1f%% ka matlab kya hai?
Ye ek formatting style hai:

% = start of a format specifier

1.1f = decimal format  1 digit before decimal, 1 digit after

%% = literal percentage symbol show karna

```python
explode = [0.1,0,0,0,0]

plt.pie(slices, labels = labels,explode = explode, shadow = True, startangle = 90, autopct = "%1.1f%%", wedgeprops= {'edgecolor': 'black'})
plt.show()
```
