# What are Subplots?
Subplots are multiple plots displayed in a single figure arranged in a grid layout. They allow you to 
present several related graphs side-by-side or stacked vertically/horizontally, making it easier to compare
different datasets or visualize multiple aspects of data in one view.

# Function Signature

subplots(nrows=1, ncols=1, sharex=False, sharey=False, squeeze=True, subplot_kw=None, gridspec_kw=None, **fig_kw)


# Why are Subplots Useful?
1. Organize multiple plots in a structured grid.

2. Compare different datasets or visualizations side-by-side.

3. Save space by combining graphs in a single figure window.

4. Control shared axes for better comparison of data trends.

# Attributes of Subplots
nrows, ncols: Number of rows and columns in the subplot grid.

sharex, sharey: Whether subplots share the same x-axis or y-axis scales.

squeeze: If True, reduces dimensions of returned arrays when possible.

subplot_kw: Dictionary of keywords passed to each subplot.

gridspec_kw: Dictionary of keywords passed to the grid specification.

fig_kw: Additional keywords passed to the figure creation, like figsize.

The function returns:

fig: The overall figure object.

axes: Array or single Axes objects representing the subplots.


# When are Subplots Used?
1. When visualizing multiple related charts together.

2. When you want to compare different variables or scenarios side-by-side.

3. When plotting data that benefits from different visualizations (e.g., bar chart and scatter plot) in one figure.

4. When organizing complex visual reports or dashboards.

# Purpose of Subplots
1. To combine multiple plots into one figure cleanly.

2. To enhance comparative analysis between datasets.

3. To maximize information display within a limited space.

4. To provide a coherent visual structure for related data.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('data_1.csv')
data.head(3)
```

```python
ages = data['Age']
dev = data['All_Devs']
python = data['Python']
js = data['JavaScript']

plt.plot(ages, python, label = 'Python')
plt.plot(ages, js, label = 'JavaScript')

plt.plot(ages, dev, color = '#444444', linestyle = '--', label = 'All Devs')

plt.legend()

plt.title('Median Salary by Age')
plt.xlabel('Ages')
plt.ylabel('Salary')

plt.tight_layout()

plt.show()
```

```python
ages = data['Age']
dev = data['All_Devs']
python = data['Python']
js = data['JavaScript']

fig, ax = plt.subplots()

ax.plot(ages, python, label = 'Python')
ax.plot(ages, js, label = 'JavaScript')

ax.plot(ages, dev, color = '#444444', linestyle = '--', label = 'All Devs')

ax.legend()

ax.set_title('Median Salary by Age')
ax.set_xlabel('Ages')
ax.set_ylabel('Salary')

plt.tight_layout()

plt.show()
```
